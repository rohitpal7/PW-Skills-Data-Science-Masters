```python
#Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.
#Ans.
import scipy.stats as stats

sample_mean = 50
sample_std = 5
n = 100  # sample size

# Calculate the standard error of the mean
standard_error = sample_std / (n ** 0.5)

# Calculate the 95% confidence interval
lower_ci, upper_ci = stats.norm.interval(0.95, loc=sample_mean, scale=standard_error)

print("95% Confidence Interval: ({:.2f}, {:.2f})".format(lower_ci, upper_ci))
```

```python
#Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.
#Ans.
import numpy as np
import scipy.stats as stats

# Observed frequencies
observed_freq = np.array([18, 22, 20, 9, 11, 20])

# Expected frequencies
expected_freq = np.array([20, 20, 20, 10, 10, 20])

# Calculate the chi-square test statistic
chi2_statistic, p_value = stats.chisquare(observed_freq, expected_freq)

# Calculate the degrees of freedom
df = len(observed_freq) - 1

# Calculate the critical value at a significance level of 0.05
critical_value = stats.chi2.ppf(0.95, df)

# Compare the chi-square statistic with the critical value and interpret the results
if chi2_statistic > critical_value:
    print("The chi-square test statistic ({:.2f}) exceeds the critical value ({:.2f}), so we reject the null hypothesis.".format(chi2_statistic, critical_value))
else:
    print("The chi-square test statistic ({:.2f}) does not exceed the critical value ({:.2f}), so we fail to reject the null hypothesis.".format(chi2_statistic, critical_value))
```

```python
#Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.
#Ans.
import statsmodels.api as sm

# Calculate the confidence interval
conf_int = sm.stats.proportion_confint(60, 500, alpha=0.05, method='normal')

# Print the results
print("95% confidence interval:", conf_int)
```

```python
#Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.
#Ans.
import scipy.stats as stats

# Define the sample size, mean, and standard deviation
n = 100
mean = 75
std_dev = 12

# Calculate the standard error of the mean
sem = std_dev / (n ** 0.5)

# Calculate the confidence interval
conf_int = stats.norm.interval(0.9, loc=mean, scale=sem)

# Print the results
print("90% confidence interval:", conf_int)
```

```python
#Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.
#Ans.
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the degrees of freedom
df = 10

# Define the range of x values
x = np.linspace(0, 30, 200)

# Create the chi-square distribution
chisq = stats.chi2.pdf(x, df)

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, chisq, 'b-', lw=2, label='Chi-Square PDF')

# Shade the area corresponding to a chi-square statistic of 15
ax.fill_between(x, chisq, where=x>=15, color='gray', alpha=0.5)

# Label the axes
ax.set_xlabel('Chi-Square Statistic')
ax.set_ylabel('Probability Density')

# Add a legend
ax.legend(loc='best')

# Show the plot
plt.show()
```

```python
#Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.
#Ans.
import numpy as np
import scipy.stats as stats

# Define the sample size and proportion
n = 1000
p = 520/n

# Calculate the z-score for a 99% confidence level
z = stats.norm.ppf(0.995)

# Calculate the margin of error
margin_of_error = z * np.sqrt(p*(1-p)/n)

# Calculate the confidence interval
CI = (p - margin_of_error, p + margin_of_error)

print(CI)
```

```python
#Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.
#Ans.
import scipy.stats as stats

# Observed frequencies
observed_freq = [45, 55]

# Expected frequencies assuming a fair coin
expected_freq = [50, 50]

# Perform chi-square goodness of fit test
chi_square, p_value = stats.chisquare(observed_freq, f_exp=expected_freq)

# Output results
print("Chi-square statistic:", chi_square)
print("P-value:", p_value)

# Check if null hypothesis is rejected or not
if p_value < 0.05:
    print("Reject null hypothesis - The coin is biased towards tails.")
else:
    print("Fail to reject null hypothesis - The coin is fair.")
```

```python
#Q9. A study was conducted to determine if there is an association between smoking status (smoker ornon-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below.Conduct a chi-square test for independence to determine if there is a significant association between smoking status and lung cancer diagnosis.
#Ans.
import numpy as np
import scipy.stats as stats

# Contingency table
cont_table = np.array([[60, 140],
                       [30, 170]])

# Calculate chi-square statistic and p-value
chi2_stat, p_value, dof, expected_freq = stats.chi2_contingency(cont_table)

# Output results
print("Chi-square statistic:", chi2_stat)
print("P-value:", p_value)

# Check if null hypothesis is rejected or not
if p_value < 0.05:
    print("Reject null hypothesis - There is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("Fail to reject null hypothesis - There is no significant association between smoking status and lung cancer diagnosis.")
```

```python
#Q10. A study was conducted to determine if the proportion of people who prefer milk chocolate, dark chocolate, or white chocolate is different in the U.S. versus the U.K. A random sample of 500 people from the U.S. and a random sample of 500 people from the U.K. were surveyed. The results are shown in the contingency table below. Conduct a chi-square test for independence to determine if there is a significant association between chocolate preference and country of origin.
#Ans.
import numpy as np
import scipy.stats as stats

# Contingency table
cont_table = np.array([[200, 150, 150],
                       [225, 175, 100]])

# Calculate chi-square statistic and p-value
chi2_stat, p_value, dof, expected_freq = stats.chi2_contingency(cont_table)

# Output results
print("Chi-square statistic:", chi2_stat)
print("P-value:", p_value)

# Check if null hypothesis is rejected or not
if p_value < 0.01:
    print("Reject null hypothesis - There is a significant association between chocolate preference and country of origin.")
else:
    print("Fail to reject null hypothesis - There is no significant association between chocolate preference and country of origin.")
```

```python
#Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.
#Ans.
import scipy.stats as stats

# Sample size
n = 30

# Sample mean
sample_mean = 72

# Sample standard deviation
sample_std = 10

# Hypothesized mean
hypo_mean = 70

# Degrees of freedom
df = n - 1

# Calculate t-statistic
t = (sample_mean - hypo_mean) / (sample_std / (n ** 0.5))

# Calculate p-value
p_value = stats.t.sf(abs(t), df) * 2

# Output results
print("t-statistic:", t)
print("p-value:", p_value)

# Check if null hypothesis is rejected or not
if p_value < 0.05:
    print("Reject null hypothesis - Population mean is significantly different from 70.")
else:
    print("Fail to reject null hypothesis - Population mean is not significantly different from 70.")
```
