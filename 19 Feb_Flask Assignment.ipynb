```python
#Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.
#Ans.
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#Q3. What is App routing in Flask? Why do we use app routes?
#Ans.
#In Flask, app routing is the process of mapping URLs to Python functions in a Flask application. 
#The main purpose of app routing is to handle incoming HTTP requests to a Flask application and direct them to the appropriate function to generate an HTTP response. 
#By using app routing, you we create URLs that map to specific views, or functions, that generate responses to those URLs. 
```

```python
#Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
#Company Name: ABC Corporation
#Location: India
#Contact Detail: 999-999-9999
#Ans.
from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return 'Welcome to ABC Corporation'

@app.route("/")
def details():
    return 'Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999'

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
#Ans.
#In Flask, the url_for function is used for URL building. It takes the endpoint name of a view function as its first argument and returns the URL for that endpoint. 
#This is useful when you want to create links between different pages or pass dynamic parameters to a URL.
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'This is the home page. <a href="{}">Go to about page</a>'.format(url_for('about'))

@app.route('/about')
def about():
    return 'This is the about page. <a href="{}">Go to home page</a>'.format(url_for('index'))

if __name__ == '__main__':
   app.run(host="0.0.0.0")
```
