```python
#Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the dataset have?
#Ans.
import pandas as pd

# Load the flight price dataset
df = pd.read_csv('flight_price.csv')

# Examine the dimensions of the dataset
print('Number of rows: ', df.shape[0])
print('Number of columns: ', df.shape[1])
```

```python
#Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.
#Ans.
import pandas as pd
import matplotlib.pyplot as plt

# Load the flight price dataset
df = pd.read_csv('flight_price.csv')

# Create a histogram of the flight prices
plt.hist(df['Price'], bins=20)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Q3. What is the range of prices in the dataset? What is the minimum and maximum price?
#Ans.
import pandas as pd

# Load the flight price dataset
df = pd.read_csv('flight_price.csv')

# Get the range, minimum, and maximum prices
range_price = df['Price'].max() - df['Price'].min()
min_price = df['Price'].min()
max_price = df['Price'].max()

print("Range of prices:", range_price)
print("Minimum price:", min_price)
print("Maximum price:", max_price)
```

```python
#Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.
#Ans.
import pandas as pd
import matplotlib.pyplot as plt

# Load the flight price dataset
df = pd.read_csv('flight_price.csv')

# Create a boxplot of price by airline
plt.figure(figsize=(10,6))
plt.boxplot([df[df['Airline']=='IndiGo']['Price'], 
             df[df['Airline']=='Jet Airways']['Price'], 
             df[df['Airline']=='Air India']['Price'], 
             df[df['Airline']=='SpiceJet']['Price'], 
             df[df['Airline']=='Multiple carriers']['Price'],
             df[df['Airline']=='GoAir']['Price'],
             df[df['Airline']=='Vistara']['Price'],
             df[df['Airline']=='Air Asia']['Price']])
plt.xticks([1,2,3,4,5,6,7,8], ['IndiGo', 'Jet Airways', 'Air India', 'SpiceJet', 'Multiple carriers', 'GoAir', 'Vistara', 'Air Asia'], rotation=45)
plt.title('Boxplot of Flight Prices by Airline')
plt.ylabel('Price (in INR)')
plt.show()
```
