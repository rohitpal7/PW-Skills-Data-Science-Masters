```python
#Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?
#Ans.
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Load the data
data = pd.read_csv('wine_quality.csv')

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Perform PCA
pca = PCA()
pca.fit(data_scaled)

# Calculate the explained variance ratio
explained_var_ratio = pca.explained_variance_ratio_

# Determine the minimum number of principal components
cumulative_var_ratio = 0
num_components = 0

for var_ratio in explained_var_ratio:
    cumulative_var_ratio += var_ratio
    num_components += 1
    if cumulative_var_ratio >= 0.9:
        break

print(f"The minimum number of principal components required to explain 90% of the variance in the data is {num_components}.")
```
