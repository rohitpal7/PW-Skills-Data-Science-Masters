```python
#Q2. What is the objective function of a linear SVM?
#Ans.
#The objective function of a linear SVM (Support Vector Machine) is to maximize the margin between the hyperplane and the closest data points of each class, while also minimizing the classification error. 
#The objective function of a linear SVM can be expressed as:
min ||w||^2 + C * sum(max(0, 1 - y_i(w^T x_i + b)))
```

```python
#Q3. What is the kernel trick in SVM?
#Ans.
#The kernel trick in SVM (Support Vector Machine) is a mathematical technique that allows us to implicitly transform the input feature space into a higher-dimensional space, without actually computing the coordinates of the data points in that space. 
#This allows us to apply linear methods in the high-dimensional space, while only working with the original data points in the low-dimensional input feature space.
#The most commonly used kernel functions in SVM are the linear kernel, polynomial kernel, Gaussian RBF kernel, and sigmoid kernel. 
#The choice of kernel function depends on the nature of the data and the specific problem being addressed.
```

```python
#Q6. SVM Implementation through Iris dataset.
#Ans.
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# Load the iris dataset
iris = load_iris()

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=42)

# Train a linear SVM classifier on the training set
clf = LinearSVC(C=1, loss="hinge", random_state=42)
clf.fit(X_train, y_train)

# Predict the labels for the testing set
y_pred = clf.predict(X_test)

# Compute the accuracy of the model on the testing set
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Plot the decision boundaries of the trained model using two of the features
fig, ax = plt.subplots()

# Plot the training points
for i, color in zip(range(3), ["r", "b", "g"]):
    idx = np.where(y_train == i)
    ax.scatter(X_train[idx, 0], X_train[idx, 1], c=color, label=iris.target_names[i],
               alpha=0.8, edgecolors='none')

# Plot the decision boundaries
x_min, x_max = X_train[:, 0].min() - 0.5, X_train[:, 0].max() + 0.5
y_min, y_max = X_train[:, 1].min() - 0.5, X_train[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
ax.contourf(xx, yy, Z, alpha=0.4)

# Add axis labels and legend
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])
ax.legend()
plt.show()
```
